```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# defino funciones para llamarlas luego

def check_nan(df: pd.DataFrame) -> None:
    
    """Recibe un dataframe y enseña el % de valores nulos y lo grafica"""
    
    nan_cols = df.isna().mean() * 100  # porcentaje de nulo en cada columna
    
    display(f'N nan cols: {len(nan_cols[nan_cols>0])}')
    display(nan_cols[nan_cols>0])
    
    
    # grafico de nulos en el dataframe
    plt.figure(figsize=(10, 6))     # inicia la figura y establece el tamaño
    sns.heatmap(df.isna(),          # datos
                yticklabels=False,  # quita las etiquetas del eje y
                cmap='viridis',     # mapa de color
                cbar=False,         # sin barra lateral
               )


    plt.show();
```

```python
def time_of_day(hour): # funcion para agrupar y asi reducir la cantidad de horas del dia en periodos preestablecidos
    
    if '07' in hour or '08' in hour or '09' in hour or '10' in hour or '11' in hour or '12' in hour:
        return 'Morning'
    elif '13' in hour or '14' in hour or '15' in hour or '16' in hour or '17' in hour or '18' in hour or '19' in hour or '20' in hour or '21' in hour:
        return 'Afternoon'
    elif '22' in hour or '23' in hour or '00' in hour or '01' in hour or '02' in hour or '03' in hour or '04' in hour or '05' in hour or '06' in hour:
        return 'Night'
    else:
        return 'No information'

data['unnamed: 23'] = data['time'].apply(time_of_day)
```

```python
def age_of_victims(age): # funcion para agrupar y asi reducir la cantidad de edades distintas en grupos por edad
        
        if age == 0:

            return 'No information'
    
        if age <= 12:
        
            return 'kid'
       
        elif age <= 19:

            return 'teen'
    
        elif age <= 60:

            return 'adult'
    
        elif age >= 61:

            return 'elderly'
    
        else:
            return 'No information'

data['unnamed: 22'] = data['age'].apply(age_of_victims)
```

```python
def years_to_century(year): # funcion para agrupar años por centurias
        
        if year <= 1499:

            return 'No concrete information'
    
        if  year <= 1599:
        
            return 'S.XVI'
       
        elif year <= 1699:

            return 'S.XVII'
    
        elif year <= 1799:

            return 'S.XVIII'
    
        elif year >= 1899:

            return 'S.XIX'
        
        elif year >= 1999:
             
            return 'S.XX'
        
        elif year >= 2099:

            return 'S.XXI'
        
        else:

            return 'No information'

data['case number.1'] = data['year'].apply(years_to_century)
```

```python
data.rename(columns={'case number.1': 'Century'}, inplace=True) # cambio los nombres de las columnas para darle uso con data agrupada
```
